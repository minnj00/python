## 3.2 random

````python
import random
````

````python
 random.random() #: 랜덤의 소수를 출력   
````

````python
random.randint(1, 10) # 랜덤으로 정수를 출력
````

````python
# seed
random.seed(1) 
# 우리가 봤을 때는 랜덤으로 보이지만 컴퓨터는 그 값도 계산을 해서 출력하는 것임

random.random()
````

````python
# shuffle
a = [1, 2, 3, 4, 5]
random.shuffle(a)
print(a)
````

````python
# choice 
a = ['가위', '바위', '보']
print(random.choice(a))
````

````python
# sample
a = range(1, 46)
random.sample(a, 6)
````
